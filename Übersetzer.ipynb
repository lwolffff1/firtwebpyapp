```python
 from transformers import MarianMTModel, MarianTokenizer
 from gtts import gTTS
 import os

 from IPython.display import Audio
 model_name = "Helsinki-NLP/opus-mt-de-vi"
 tokenizer = MarianTokenizer.from_pretrained(model_name)
 model = MarianMTModel.from_pretrained(model_name)

 ## Text needs to be translated

 text = input("Eingabe auf Deutsch: ")

#model requires list, so change into list
 text_list =[text]

 #Tokenize and translate
 tokens = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
 translation = model.generate(**tokens)
 translated_text = tokenizer.batch_decode(translation, skip_special_tokens=True)

 print("Übersetzt: ", translated_text[0])

 # Output voice
 tts = gTTS(translated_text[0], lang='vi')
 tts.save("output.mp3")

# Phát âm thanh trực tiếp trong Google Colab
Audio("output.mp3", autoplay=True)
```
